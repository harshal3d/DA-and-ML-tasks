```python
from matplotlib import pyplot as pt
import pandas as pd
import seaborn as sns
from scipy import stats as st
import numpy as np
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
# making new features
df["Class_compliment"]=(df["Class"]+1)%2
df["fraud amt"]=df["Amount"]*df["Class"]
df["normal amt"]=df["Amount"]*df["Class_compliment"]
df[0:20]
```

```python
fraudcount=sum(df["Class"])
normalcount=sum(df["Class_compliment"])
fraudamt=sum(df["fraud amt"])
normalamt=sum(df["normal amt"])
print('fraud count = '+str(fraudcount))
print('normal count = '+str(normalcount))
rows=fraudcount+normalcount
print('rows count = '+str(rows))
```

```python
avgfraudamt=fraudamt/fraudcount
avgnormalamt=normalamt/normalcount

print('Avg fraud amount = '+str(avgfraudamt))
print('Avg normal amount = '+str(avgnormalamt)+'\n')
max_normal_amt=max(df["normal amt"])
min_normal_amt=min(df["normal amt"])

max_frd_amt=max(df["fraud amt"])
min_frd_amt=min(df["fraud amt"])
print('max normal amount = '+str(max_normal_amt))
print('max fraud amount = '+str(max_frd_amt))
```

```python
timegap=[]
time=df["Time"]
isfrd=df["Class"]

for i in range(rows):
    if(isfrd[i]==1):
        timegap.append(time[i])

        
gaps_count = len(timegap)
i=gaps_count-1
while i>1:
    timegap[i]=timegap[i]-timegap[i-1]
    i=i-1

timegap[1:10]
```

```python
dt =pt.hist(timegap,bins=30)
# we can observe the destribution of time gap ressembels exponential 
# impling it is has memory less property 
# which further conforms the intution that each fraud is independent of others statistically
import scipy
x=scipy.mean(timegap)
print("mean of timegaps = "+str(x))
```
